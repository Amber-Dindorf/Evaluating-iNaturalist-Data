```python
Sciname = inat['verbatimScientificName'].unique()
```

```python
sciname = list(Sciname)
```

```python
type(sciname)
```

```python
from bs4 import BeautifulSoup as soup
```

```python
import time
```

```python
Common_name = []

for species in sciname:
    wiki_url = 'https://en.wikipedia.org/wiki/'+species
    name = requests.get(wiki_url).text
    name_soup = soup(name, 'html.parser')
    com_name = name_soup.select('title')[0].text[:-12]
    Common_name.append(com_name)

    time.sleep(.1)
    
```

```python
Common_name
```

```python
Common_name[:10]
```

```python
names = pd.DataFrame({'Scientific_name':sciname, 'Common_Name':Common_name})
```

```python
names.head(4)
```

```python
inat = inat.merge(names, how = 'left', left_on = 'verbatimScientificName', right_on = 'Scientific_name')
```
